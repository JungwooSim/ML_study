## 2. Exploratory Data Ananysis
### 2.1 survival_time

```python
target_gb_survival_time = train_label.groupby('survival_time').count()
```

```python
target_gb_survival_time['survival_time_percent'] = target_gb_survival_time['acc_id'] / 40000 * 100
target_gb_survival_time = target_gb_survival_time.sort_values(by='survival_time_percent', ascending=False)
target_gb_survival_time['cumulative_survival_time_percent'] = target_gb_survival_time['survival_time_percent'].cumsum()
```

### 2.2 amount_spent

```python
train_label.sort_values(by='amount_spent', ascending=False)
```

```python
print("결제금액 없는 유저 수 {} : , 전체 중 퍼센트 : {}".
    format(
        train_label[train_label['amount_spent'] == 0]['amount_spent'].count(),
        train_label[train_label['amount_spent'] == 0]['amount_spent'].count() / 40000 * 100
    )
)
```

```python
# amount_spent != 0 빈도 변수 생성
survival_time_frequency = train_label[train_label['amount_spent'] != 0][['acc_id','survival_time']].groupby('survival_time').count().reset_index()
survival_time_frequency = survival_time_frequency.rename(columns={"acc_id" : "not_zero_amount_spent_Fq"})
# amount_spent == 0 빈도 변수 생성
zero_survival_time_df = train_label[train_label['amount_spent'] == 0][['acc_id','survival_time']].groupby('survival_time').count().reset_index()
zero_survival_time_df = zero_survival_time_df.rename(columns={"acc_id" : "zero_amount_spent_Fq"})
survival_time_frequency = pd.merge(survival_time_frequency, zero_survival_time_df, on="survival_time")
# amount_spent 빈도 percent 변수 생성
survival_time_frequency['not_zero_amount_spent_percent'] = survival_time_frequency['not_zero_amount_spent_Fq'] / (survival_time_frequency['not_zero_amount_spent_Fq'] + survival_time_frequency['zero_amount_spent_Fq'])
survival_time_frequency['zero_amount_spent_percent'] = 1 - survival_time_frequency['not_zero_amount_spent_percent']
```

```python
survival_time_frequency.head()
```

```python
for i in range(0,4):
    row = i * 16
    survival_time_frequency.set_index('survival_time')[['not_zero_amount_spent_percent','zero_amount_spent_percent']][row:row+16].plot(
        kind="bar",
        stacked=True,
        figsize=(10,5)
    )
```
