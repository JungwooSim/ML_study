# 1. 도입(라이브러리, 데이터 로딩)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn')
sns.set(font_scale=2.5)

import missingno as msno
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
import scipy as sp

import warnings
warnings.filterwarnings('ignore')
from collections import OrderedDict
     
import os
%matplotlib inline

pd.set_option('display.max_columns', 150)
pd.set_option('display.max_rows', 500)
pd.options.display.float_format = '{:.2f}'.format
```

```python
os.chdir('..') #상위디렉터리로 이동
path_dir = os.getcwd()+"/champions/"
file_list = sorted(os.listdir(path_dir))

data_dict = OrderedDict()
for file in file_list:
    if file.find('csv') is not -1:
        if file.split('_')[1] == 'activity.csv':
            data_dict[file.split('.')[0]] = pd.read_csv(path_dir+file)
```

## 1.1 자료 구분값 추가(load 파일 구분)

```python
data_dict['train_activity']['kind'] = 'train'
data_dict['test1_activity']['kind'] = 'test1'
data_dict['test2_activity']['kind'] = 'test2'
data_dict['train_activity']['kind_int'] = 1
data_dict['test1_activity']['kind_int'] = 2
data_dict['test2_activity']['kind_int'] = 3

activity_all = data_dict['train_activity'].append(data_dict['test1_activity'])
del(data_dict['train_activity'])
del(data_dict['test1_activity'])
activity_all = activity_all.append(data_dict['test2_activity'])
del(data_dict['test2_activity'])
```

# 2. 탐색적 자료 분석

## 2.1 int, float64 자료형 변수 train, test1, test2 분포 확인

```python
f = len(activity_all.select_dtypes(include={'int64','float64'}).keys())
ax = len(activity_all['kind'].unique())

f, ax = plt.subplots(f, ax,figsize=(40,150))
row = 0
for activity_all_keys in activity_all.select_dtypes(include={'int64','float64'}).keys():
    col = 0
    for type_keys in activity_all['kind'].unique():
        sns.distplot(
            activity_all[activity_all['kind'] == type_keys][activity_all_keys],
            ax=ax[row][col],
            kde_kws={"label":type_keys}
        )
        col = col+1
    row = row+1
```

## 2.2 int, float64 자료형 변수 train, test1, test2 통계치 확인

```python
activity_all[activity_all.select_dtypes(include={'int64','float64'}).keys()].groupby('kind_int').describe().T
```

# 3. Feature Engineering

**FE를 통해 추가한 변수 리스트**

* **game_money_change** : 아데나 변동량
* **activity_row_count** : 열 개수
* 케릭터 개수와 리스트
    * **charid_sum_count** : 개수
    * **charid_sum_detail** : 리스트
* 활동중인 서버, 서버리스트
    * **server_sum_count** : 개수
    * **server_sum_detail** : 리스트
* 기록된 day 변수의 개수와 리스트
    * **day_sum_count** : 개수
    * **day_sum_detail** : 리스트
* 데이터 로드한 위치(train, test1, test2)의 개수와 리스트
    * **kind_sum_count** : 개수
    * **kind_sum_detail** : 리스트
* **zero_cnt_activity** : 열에서 0값을 가진 개수
