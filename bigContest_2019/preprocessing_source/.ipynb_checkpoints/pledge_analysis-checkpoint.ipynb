# 1. 도입(라이브러리, 데이터 로딩)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn')
sns.set(font_scale=2.5)

import missingno as msno
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
import scipy as sp

import warnings
warnings.filterwarnings('ignore')
from collections import OrderedDict
     
import os
%matplotlib inline

pd.set_option('display.max_columns', 150)
pd.set_option('display.max_rows', 500)
pd.options.display.float_format = '{:.2f}'.format
```

```python
os.chdir('..') #상위디렉터리로 이동
path_dir = os.getcwd()+"/champions/"
file_list = sorted(os.listdir(path_dir))

data_dict = OrderedDict()
for file in file_list:
    if file.find('csv') is not -1:
        if file.split('_')[1] == 'pledge.csv':
            data_dict[file.split('.')[0]] = pd.read_csv(path_dir+file)
```

## 1.1 자료 구분값 추가(load 파일 구분)

```python
data_dict['train_pledge']['kind'] = 'train'
data_dict['test1_pledge']['kind'] = 'test1'
data_dict['test2_pledge']['kind'] = 'test2'
data_dict['train_pledge']['kind_int'] = 1
data_dict['test1_pledge']['kind_int'] = 2
data_dict['test2_pledge']['kind_int'] = 3

pledge_all = data_dict['train_pledge'].append(data_dict['test1_pledge'])
del(data_dict['train_pledge'])
del(data_dict['test1_pledge'])
pledge_all = pledge_all.append(data_dict['test2_pledge'])
del(data_dict['test2_pledge'])
```

# 2. 탐색적 자료 분석

## 2.1 int, float64 자료형 변수 train, test1, test2 분포 확인

```python
f = len(pledge_all.select_dtypes(include={'int64','float64'}).keys())
ax = len(pledge_all['kind'].unique())

f, ax = plt.subplots(f, ax,figsize=(40,150))
row = 0
for pledge_all_keys in pledge_all.select_dtypes(include={'int64','float64'}).keys():
    col = 0
    for type_keys in pledge_all['kind'].unique():
        sns.distplot(
            pledge_all[pledge_all['kind'] == type_keys][pledge_all_keys],
            ax=ax[row][col],
            kde_kws={"label":type_keys}
        )
        col = col+1
    row = row+1
```

## 2.2 int, float64 자료형 변수 train, test1, test2 통계치 확인

```python
pledge_all[pledge_all.select_dtypes(include={'int64','float64'}).keys()].groupby('kind_int').describe().T
```

* play_char_cnt, temp_cnt, combat_char_cnt 변수에서 차이가 보인다.

# 3. Feature Engineering

**FE를 통해 추가한 변수 리스트**

* 기록된 day 변수의 개수와 리스트
    * **day_concat_count** : 개수
    * **day_concat_detail** : 리스트
* 활동중인 서버, 서버리스트
    * **server_concat_count** : 개수
    * **server_concat_detail** : 리스트
* 케릭터 개수와 리스트
    * **charid_concat_count** : 개수
    * **charid_concat_detail** : 리스트
* 가입한 혈맹 수와 리스트
    * **pledgeid_concat_count** : 개수
    * **pledgeid_concat_detail** : 리스트
* 데이터 로드한 위치(train, test1, test2)의 개수와 리스트
    * **kind_concat_count** : 개수
    * **kind_concat_detail** : 리스트
* **pledge_row_count** : 열 개수
* **play_char_mean** : 게임에 접속한 혈맹원의 평균 수
* **combat_char_mean** : 전투에 참여한 혈맹원의 평균 수
