# 1. 도입(라이브러리, 데이터 로딩)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn')
sns.set(font_scale=2.5)

import missingno as msno
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
import scipy as sp

import warnings
warnings.filterwarnings('ignore')
from collections import OrderedDict
     
import os
%matplotlib inline

pd.set_option('display.max_columns', 150)
pd.set_option('display.max_rows', 500)
pd.options.display.float_format = '{:.2f}'.format
```

```python
os.chdir('..') #상위디렉터리로 이동
path_dir = os.getcwd()+"/champions/"
file_list = sorted(os.listdir(path_dir))

data_dict = OrderedDict()
for file in file_list:
    if file.find('csv') is not -1:
        if file.split('_')[1] == 'combat.csv':
            data_dict[file.split('.')[0]] = pd.read_csv(path_dir+file)
```

## 1.1 자료 구분값 추가(load 파일 구분)

```python
data_dict['train_combat']['kind'] = 'train'
data_dict['test1_combat']['kind'] = 'test1'
data_dict['test2_combat']['kind'] = 'test2'
data_dict['train_combat']['kind_int'] = 1
data_dict['test1_combat']['kind_int'] = 2
data_dict['test2_combat']['kind_int'] = 3

combat_all = data_dict['train_combat'].append(data_dict['test1_combat'])
del(data_dict['train_combat'])
del(data_dict['test1_combat'])
combat_all = combat_all.append(data_dict['test2_combat'])
del(data_dict['test2_combat'])
```

# 2. 탐색적 자료 분석

## 2.1 class

```python
combat_all.groupby('class')['acc_id'].count().plot.bar()
```

```python
combat_class = pd.DataFrame(
    data = {
        "class" : combat_all.groupby('class')['acc_id'].count().index.tolist(),
        "class_name" : ['군주','기자','요정','마법사','다크엘프','용기사','환술사','전사'],
        "class_count" : combat_all.groupby('class')['acc_id'].count().tolist()
    }
)
combat_class['percent'] = combat_class['class_count'] / len(combat_all['class'])
```

```python
combat_class.sort_values(by="class_count", ascending=False)
```

## 2.2 level

```python
combat_all.groupby('level')['acc_id'].count().sort_values(ascending=False).plot.bar()
```

```python
combat_level = pd.DataFrame(
    data = {
        "level" : combat_all.groupby('level')['acc_id'].count().index.tolist(),
        "level_count" : combat_all.groupby('level')['acc_id'].count().tolist()
    }
)
combat_level['percent'] = combat_level['level_count'] / len(combat_all['level'])
```

```python
combat_level.sort_values(by="percent", ascending=False)
```

* 많은 부분이 level 16단계에 있는것을 알 수 있다.

## 2.3 int, float64 자료형 변수 train, test1, test2 분포 확인

```python
f = len(combat_all.select_dtypes(include={'int64','float64'}).keys())
ax = len(combat_all['kind'].unique())

f, ax = plt.subplots(f, ax,figsize=(40,150))
row = 0
for combat_all_keys in combat_all.select_dtypes(include={'int64','float64'}).keys():
    col = 0
    for type_keys in combat_all['kind'].unique():
        sns.distplot(
            combat_all[combat_all['kind'] == type_keys][combat_all_keys],
            ax=ax[row][col],
            kde_kws={"label":type_keys}
        )
        col = col+1
    row = row+1
```

# 3. Feature Engineering

**FE를 통해 추가한 변수 리스트**

* 가지고 있는 케릭터 클래스와 클래스 수
    * **class_concat_count** : 개수
    * **class_concat_detail** : 리스트
* 케릭터 개수와 리스트
    * **charid_concat_count** : 개수
    * **charid_concat_detail** : 리스트
* 활동중인 서버, 서버리스트
    * **server_concat_count** : 개수
    * **server_concat_detail** : 리스트
* 가지고 있는 케릭터 레벨과 리스트
    * **level_concat_count** : 개수
    * **level_concat_detail** : 리스트
* 기록된 day 변수의 개수와 리스트
    * **day_concat_count** : 개수
    * **day_concat_detail** : 리스트
* 데이터 로드한 위치(train, test1, test2)의 개수와 리스트
    * **kind_concat_count** : 개수
    * **kind_concat_detail** : 리스트
* **level_mean** : 가지고 있는 케릭터의 평균 레벨
* **level_max** : 가지고 있는 케릭터의 평균 레벨
* **combat_row_count** : 열 개수
* **zero_cnt_combat** : 열에서 0값을 가진 개수
